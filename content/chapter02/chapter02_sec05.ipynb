## Das erste Diagramm

Wir starten direkt mit einem Beispiel. Das gedämpfte Schwingungssignal aus
Kapitel 2.1 soll als Linienplot dargestellt werden:

````
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('seaborn-v0_8')

# Schwingungssignal (wie in Kapitel 2.1)
A     = 5.0
delta = 1.5
f     = 10.0
t     = np.linspace(0, 2, 1000)
a     = A * np.exp(-delta * t) * np.sin(2 * np.pi * f * t)

# Diagramm erzeugen
fig, ax = plt.subplots()
ax.plot(t, a)
plt.show()
````

Mit drei Zeilen haben wir ein vollständiges Diagramm. Wir sehen das typische
Bild einer gedämpften Schwingung: hohe Amplitude am Anfang, die mit der Zeit
abnimmt.

Schauen wir uns die drei Zeilen genauer an. `plt.subplots()` erzeugt zwei
Objekte: eine **Figure** und eine **Axes**. Die Figure ist das äußere Fenster,
die Axes ist der eigentliche Zeichenbereich mit Achsen, Kurven und
Beschriftungen. Das Muster `fig, ax = plt.subplots()` werden wir in jedem
Diagramm verwenden.

`ax.plot(t, a)` zeichnet eine Linie, die die Punkte `(t[0], a[0])`,
`(t[1], a[1])` usw. verbindet. Das erste Argument ist immer die x-Achse,
das zweite die y-Achse. `plt.show()` zeigt das fertige Diagramm an.

**Hinweis: Was ist der Unterschied zwischen Figure und Axes?**
In Matplotlib gibt es eine klare Hierarchie: Eine **Figure** kann mehrere
**Axes** enthalten. Eine Axes ist ein einzelner Zeichenbereich mit zwei
Achsen (x und y). Das klingt verwirrend, weil "Axes" im Englischen sowohl
"Achsen" als auch den Zeichenbereich meinen kann. Als Faustregel gilt: Alles,
was das Aussehen einer einzelnen Grafik betrifft (Kurven, Beschriftungen,
Gitter), steuern wir über das `ax`-Objekt. Alles, was die gesamte Abbildung
betrifft (Größe, Speichern), steuern wir über `fig`.

## Beschriftungen, Titel und Gitter

Ein Diagramm ohne Achsenbeschriftungen ist in der Ingenieurpraxis wertlos.
Wir ergänzen unser Diagramm um alle notwendigen Elemente:

````
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(t, a)

ax.set_xlabel('Zeit in s')
ax.set_ylabel('Beschleunigung in m/s²')
ax.set_title('Gedämpfte Schwingung eines Maschinenlagers')
ax.legend(['Sensor 1'])
ax.grid(True)

plt.show()
````

`figsize=(8, 4)` legt die Breite und Höhe der Figure in Zoll fest. Mit
`set_xlabel()` und `set_ylabel()` beschriften wir die Achsen, mit `set_title()`
setzen wir einen Titel. `legend()` zeigt eine Legende an, `grid(True)` fügt
ein Gitter hinzu.

Wir können auch den Linienstil und die Farbe anpassen. Matplotlib kennt eine
Vielzahl von Farben und Linienstilen, die wir als Argumente an `ax.plot()`
übergeben:

````
fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(t, a, color='red', linewidth=1.5, linestyle='solid')

ax.set_xlabel('Zeit in s')
ax.set_ylabel('Beschleunigung in m/s²')
ax.set_title('Gedämpfte Schwingung eines Maschinenlagers')
ax.grid(True)

plt.show()
````

Die wichtigsten Argumente für `ax.plot()` sind `color` für die Farbe,
`linewidth` für die Linienbreite und `linestyle` für den Linienstil. Mögliche
Werte für `linestyle` sind `'solid'`, `'dashed'`, `'dotted'` und
`'dashdot'`.

## Mehrere Kurven in einem Diagramm

Oft wollen wir mehrere Signale vergleichen. Dazu rufen wir `ax.plot()` einfach
mehrfach auf. Jede Kurve bekommt ein `label`, das dann in der Legende erscheint:

````
# Zweites Signal: Oberschwingung mit doppelter Frequenz
a2 = (A / 2) * np.exp(-delta * t) * np.sin(2 * np.pi * 2 * f * t)

fig, ax = plt.subplots(figsize=(8, 4))

ax.plot(t, a,  color='#4C72B0', linewidth=1.5, label='Grundschwingung (10 Hz)')
ax.plot(t, a2, color='#DD8452', linewidth=1.5, label='Oberschwingung (20 Hz)')

ax.set_xlabel('Zeit in s')
ax.set_ylabel('Beschleunigung in m/s²')
ax.set_title('Grundschwingung und Oberschwingung')
ax.legend()
ax.grid(True)

plt.show()
````

Das `label`-Argument in `ax.plot()` und der anschließende Aufruf von
`ax.legend()` ohne Argumente reicht aus, damit Matplotlib die Legende
automatisch befüllt. Das ist sauberer als eine Liste von Strings an
`ax.legend()` zu übergeben, weil die Beschriftung direkt beim jeweiligen
`ax.plot()`-Aufruf steht.

## Diagramm speichern mit `fig.savefig()`

In der Praxis wollen wir Diagramme nicht nur anzeigen, sondern auch in
Berichten verwenden. `fig.savefig()` speichert das Diagramm als Datei:

````
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(t, a, label='Sensor 1')
ax.set_xlabel('Zeit in s')
ax.set_ylabel('Beschleunigung in m/s²')
ax.set_title('Gedämpfte Schwingung')
ax.legend()
ax.grid(True)

fig.savefig('schwingung.png', dpi=150, bbox_inches='tight')
plt.show()
````

Das Argument `dpi=150` legt die Auflösung in Punkten pro Zoll fest. Für
Berichte und Präsentationen sind 150 bis 300 dpi üblich. `bbox_inches='tight'`
sorgt dafür, dass Beschriftungen am Rand nicht abgeschnitten werden.
Matplotlib erkennt das Dateiformat automatisch an der Dateiendung: `.png`
für Rastergrafiken, `.pdf` oder `.svg` für Vektorgrafiken.

## Zusammenfassung und Ausblick

Wir haben Matplotlib als Visualisierungswerkzeug kennengelernt. Das Muster
`fig, ax = plt.subplots()` bildet die Grundlage jedes Diagramms. Mit
`ax.plot()` zeichnen wir Linien, mit `set_xlabel()`, `set_ylabel()`,
`set_title()`, `legend()` und `grid()` beschriften und gestalten wir das
Diagramm. Mehrere Kurven lassen sich durch wiederholte `ax.plot()`-Aufrufe
kombinieren, `fig.savefig()` speichert das Ergebnis als Datei.

Im nächsten Kapitel erweitern wir unsere Diagramme um **Subplots**, also
mehrere Zeichenbereiche in einer Figure. Das brauchen wir zum Beispiel, um
Strom, Drehzahl und Drehmoment eines Motors gleichzeitig darzustellen.
