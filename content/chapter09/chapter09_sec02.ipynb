def trapez_paarweise(t, v):
    n   = len(v)
    dt  = t[1] - t[0]
    summe = 0.0
    for i in range(0, n - 1, 2):
        summe += 0.5 * (v[i] + v[i + 1]) * dt
    return summe

t_a = np.array([0.0, 1.0, 2.0, 3.0, 4.0])
